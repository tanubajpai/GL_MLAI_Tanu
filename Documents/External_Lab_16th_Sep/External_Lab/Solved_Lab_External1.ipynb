## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa
      -- Iris Versicolour
      -- Iris Virginica

### Find out the mean and variance for each column but for class column 

```python
# Changing variables from object to float

Iris_Copy1['Sepal Length (in cm)'] =Iris_Copy1['Sepal Length (in cm)'].astype('float')
Iris_Copy1['Sepal Width in (cm)'] =Iris_Copy1['Sepal Width in (cm)'].astype('float')

Iris_Copy1['Petal length (in cm)'] =Iris_Copy1['Petal length (in cm)'].astype('float')
Iris_Copy1['Petal width (in cm)'] =Iris_Copy1['Petal width (in cm)'].astype('float')
Iris_Copy1['SepalExtra'] =Iris_Copy1['SepalExtra'].astype('float')



# Calculating mean

Iris_Copy1['Sepal Length (in cm)'].mean()
Iris_Copy1['Sepal Width in (cm)'].mean()
Iris_Copy1['Petal length (in cm)'].mean()
Iris_Copy1['Petal width (in cm)'].mean()
Iris_Copy1['SepalExtra'].mean()

# Calculating Variance
Iris_Copy1['Sepal Length (in cm)'].var()
Iris_Copy1['Sepal Width in (cm)'].var()
Iris_Copy1['Petal length (in cm)'].var()
Iris_Copy1['Petal width (in cm)'].var()
Iris_Copy1['SepalExtra'].var()
 
```

### Write a function that accepts two numbers as input and prints them - Pass the Sepal length and sepal width of 5th row and print the output

```python
def add(nums): #nums can have any number of arguments
    
    sum_value = 0
    
    for num in nums:
        sum_value = sum_value + num
    
    return sum_value

# Sepal length of 5th row & Sepal width of 5th row
Iris_Copy_Slice5 = Iris_Copy.iloc[5, [0,1]]
Iris_Copy_Slice5


#Numerical value of 5th observation of  Sepal length is 5.4 & Sepal Width is 3.9

# Using the above function and getting output
print(add([5.4,3.9]))
```
