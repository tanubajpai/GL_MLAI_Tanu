# Q.1- Part2 :  Replacing '-' with '0'

```python
# Creating copy of Laliga data

Laliga_copy= Laliga.copy()


# Using replace function for all the columns whereever '-' is present

Laliga_copy['Seasons']=Laliga_copy['Seasons'].replace('-','0')
Laliga_copy['Points']=Laliga_copy['Points'].replace('-','0')
Laliga_copy['GamesPlayed']=Laliga_copy['GamesPlayed'].replace('-','0')
Laliga_copy['GamesWon']=Laliga_copy['GamesWon'].replace('-','0')
Laliga_copy['GamesDrawn']=Laliga_copy['GamesDrawn'].replace('-','0')
Laliga_copy['GamesLost']=Laliga_copy['GamesLost'].replace('-','0')
Laliga_copy['GoalsFor']=Laliga_copy['GoalsFor'].replace('-','0')
Laliga_copy['GoalsAgainst']=Laliga_copy['GoalsAgainst'].replace('-','0')
Laliga_copy['Champion']=Laliga_copy['Champion'].replace('-','0')
Laliga_copy['Runner-up']=Laliga_copy['Runner-up'].replace('-','0')
Laliga_copy['Third']=Laliga_copy['Third'].replace('-','0')
Laliga_copy['Fourth']=Laliga_copy['Fourth'].replace('-','0')
Laliga_copy['Fifth'] =  Laliga_copy['Fifth'].replace('-','0')
Laliga_copy['Sixth'] =  Laliga_copy['Sixth'].replace('-','0')
Laliga_copy['T'] =  Laliga_copy['T'].replace('-','0')
Laliga_copy['BestPosition'] =  Laliga_copy['BestPosition'].replace('-','0')

# Seeing records
Laliga_copy.head()
```

# Q.2-     Print all the teams which have started playing between 1930-1980

```python
# Creating another copy
Laliga_copy1 = Laliga_copy.copy()


#Replacing 1929 with '1929-30' so that 'Debute' value be uniform in whole column
Laliga_copy1['Debut'] =  Laliga_copy1['Debut'].replace('1929','1929-30')


# Seeing records
Laliga_copy1.head()


#Extracting the year(firt four numbers) from 'Debute' column
Laliga_copy1['Debut'] = Laliga_copy1['Debut'].str.extract('(....)', expand=True)


# Creating another copy
Laliga_copy2 = Laliga_copy1.copy()


# Changing 'Year value' from string to Integer
Laliga_copy2['Debut'] = Laliga_copy2['Debut'].astype('int64')

# confirming if the code worked
#Laliga_copy2.info()
    

# Extracting the records for the  teams which debut between 1930-1980
Laliga_copy3 = Laliga_copy2[(Laliga_copy2.Debut >= 1930) & (Laliga_copy2.Debut <= 1980)]


# Counting the number of records falling under the above category
len(Laliga_copy3)

# Printing the final output , i.e., teams which debuted between 1930-1980
Laliga_Debut = Laliga_copy3.iloc[:,[1,17]]
Laliga_Debut.sort_values(by = 'Debut')
```
